```python
import pandas as pd
import pyarrow
import os
import time
```

```python
#Lendo o arquivo de teste no formato CSV
start = time.time()
df = pd.read_csv(
    "labeled_messages_csv.csv",
    low_memory=False
)
tempo_escrita_csv = time.time() - start
```

```python
#Convertendo o arquivo, inicialmente em CSV, em Parquet
start = time.time()
df.to_parquet(
    "labeled_messages_parquet.parquet",
    index = False,
    engine = 'pyarrow'
)
tempo_escrita_parquet = time.time() - start
```

```python
tamanho_csv = os.path.getsize(r"C:\Users\higor\Formatos de arquivos\labeled_messages_csv.csv")
tamanho_parquet = os.path.getsize(r"C:\Users\higor\Formatos de arquivos\labeled_messages_parquet.parquet")
```

```python
print("\n" + "="*40)
print("COMPARAÇÃO DE TAMANHO EM DISCO:")
print("="*40)
print(f"CSV:     {tamanho_csv / 1024 / 1024:8.2f} MB")
print(f"Parquet: {tamanho_parquet / 1024 / 1024:8.2f} MB")
print(f"Redução: {(1 - tamanho_parquet / tamanho_csv) * 100:.1f}% menor")
print("="*40)
print("\nTEMPO DE ESCRITA:")
print("="*40)
print(f"CSV:     {tempo_escrita_csv:.3f}s")
print(f"Parquet: {tempo_escrita_parquet:.3f}s")
print(f"Ganho:   {(1 - tempo_escrita_parquet / tempo_escrita_csv) * 100:.1f}% mais rápido")
print("="*40)
```
